Feature Selection


```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

X = df.iloc[:,1:31]
Y = df['diagnosis']
clf = ExtraTreesClassifier(n_estimators=50)
clf.fit(X,Y)
print(clf.feature_importances_)
feat_importances = pd.Series(clf.feature_importances_, index = X.columns)
feat_importances.nlargest(12).plot.bar()
x = feat_importances.nlargest(8)
list1 = feat_importances.keys().to_list()
X = df[['area_worst','concave points_worst','concave points_mean','perimeter_worst','radius_worst','perimeter_mean','concavity_mean','area_se']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
# predictions
rfc_predict = rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score,recall_score
confusion_matrix(y_test, rfc_predict)

```

```python
accuracy_score(y_test,rfc_predict)
```

```python
recall_score(y_test,rfc_predict, pos_label = 'M')
```

```python
f1_score(y_test,rfc_predict, pos_label = 'M')
```

Support Vector Machine

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train,y_train)
clf_predict = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score,recall_score
confusion_matrix(y_test, clf_predict)
```

```python
accuracy_score(y_test,clf_predict)
```

```python
recall_score(y_test,clf_predict, pos_label = 'M')
```

```python
f1_score(y_test,clf_predict, pos_label = 'M')
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=0).fit(x_train,y_train)
lr_predict = lr.predict(x_test)
```

```python

confusion_matrix(y_test, lr_predict)
```

```python
accuracy_score(y_test,lr_predict)
```

```python
recall_score(y_test,lr_predict, pos_label = 'M')
```

```python
f1_score(y_test,lr_predict, pos_label = 'M')
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(x_train,y_train)
neigh_predict = neigh.predict(x_test)
```
